### Simple Linear Regression Analysis 

#### Question: Does Video Views influence number of subscribers?

```python
df.head()
```

```python
# collecting x & y
X = df['video views'].values
Y = df['subscribers'].values
```

```python
# calculate mean of x & y using an inbuilt numpy method mean()
mean_x = np.mean(X)
mean_y = np.mean(Y)
```

```python
# y= c + m* X
#Calculatte m and C
```

```python
# total no.of input values
m = len(X)
```

```python
# using the formula to calculate m & c
numer = 0
denom = 0
for i in range(m):
  numer += (X[i] - mean_x) * (Y[i] - mean_y)
  denom += (X[i] - mean_x) ** 2
m = numer / denom
c = mean_y - (m * mean_x)
```

```python
print (f'm = {m} \nc = {c}')
```

```python
#Now that weve our m & c, lets plot the input points and the regression line.
```

```python
# plotting values and regression line
max_x = np.max(X) + 100
min_x = np.min(Y) - 100
```

```python
# calculating line values x and y
x = np.linspace (min_x, max_x, 100)
y = c + m * x
```

```python
plt.xlabel('Video Views')
plt.ylabel('Number of Subscribers')
plt.plot(x, y, color='#58b970', label='Regression Line')
plt.scatter(X, Y, c='#ef5423', label='data points')
```

```python
#A linear relationship exists. 
```
